# CUSTOMER SEGMENTATION MARKETING CAMPAIGN

```python
#Importing the Libraries
import numpy as np
import pandas as pd
import datetime
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import colors
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt, numpy as np
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import AgglomerativeClustering
from matplotlib.colors import ListedColormap
from sklearn import metrics
import warnings
import sys
if not sys.warnoptions:
    warnings.simplefilter("ignore")
np.random.seed(42)
```

```python
#Loading the dataset
data = pd.read_csv('marketing_campaign.csv', sep="\t")
print("Number of datapoints:", len(data))
data.head()
```

```python
#Information on features
data.info()
```

```python
#To remove the NA values
data = data.dropna()
print("The total number of data-points after removing the rows with missing values are:", len(data))
```

```python
data["Dt_Customer"] = pd.to_datetime(data["Dt_Customer"], dayfirst=True)
```

```python
# 1. Efficiently calculate the newest and oldest dates
newest_date = data["Dt_Customer"].dt.date.max()
oldest_date = data["Dt_Customer"].dt.date.min()

print("The newest customer's enrolment date in the records:", newest_date)
print("The oldest customer's enrolment date in the records:", oldest_date)
```

```python
# 2. Vectorized creation of the "Customer_For" feature

# The Series subtraction (newest_date - every date in the column) is vectorized
# This results in a Pandas Timedelta Series.
time_delta_series = newest_date - data["Dt_Customer"].dt.date
```

```python
# Convert the resulting Series into a proper Pandas Timedelta Series
time_delta_series = pd.to_timedelta(time_delta_series)
```

```python
# Extract the number of days as an integer from the Timedelta Series
data["Customer_For"] = time_delta_series.dt.days
```

```python
data["Customer_For"].head()
```

```python
print("Total categories in the feature Marital_Status:\n", data["Marital_Status"].value_counts(), "\n")
print("Total categories in the feature Education:\n", data["Education"].value_counts())
```

#FEATURE ENGINEERING

```python
#Age of customer today
data["Age"] = 2025-data["Year_Birth"]
```

```python
#Total spendings on various items
data["Spent"] = data["MntWines"]+ data["MntFruits"]+ data["MntMeatProducts"]+ data["MntFishProducts"]+ data["MntSweetProducts"]+ data["MntGoldProds"]
```

```python
#Deriving living situation by marital status"Alone"
data["Living_With"]=data["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})
```

```python
#Feature indicating total children living in the household
data["Children"]=data["Kidhome"]+data["Teenhome"]
```

```python
#Feature for total members in the household
data["Family_Size"] = data["Living_With"].replace({"Alone": 1, "Partner":2})+ data["Children"]
```

```python
#Feature pertaining parenthood
data["Is_Parent"] = np.where(data.Children> 0, 1, 0)
```

```python
#Segmenting education levels in three groups
data["Education"]=data["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})
```

```python
#For clarity
data=data.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})
```

```python
#Dropping some of the redundant features
to_drop = ["Marital_Status", "Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID"]
data = data.drop(to_drop, axis=1)
data.describe()
```

```python
#To plot some selected features

#Setting up colors preferences
sns.set(rc={"axes.facecolor":"#FFF9ED","figure.facecolor":"#FFF9ED"})
pallet = ["#682F2F", "#9E726F", "#D6B2B1", "#B9C0C9", "#9F8A78", "#F3AB60"]
cmap = colors.ListedColormap(["#682F2F", "#9E726F", "#D6B2B1", "#B9C0C9", "#9F8A78", "#F3AB60"])
```

```python
#Plotting following features
To_Plot = [ "Income", "Recency", "Customer_For", "Age", "Spent", "Is_Parent"]
print("Relative Plot Of Some Selected Features: A Data Subset")
plt.figure()
sns.pairplot(data[To_Plot], hue= "Is_Parent",palette= (["#682F2F","#F3AB60"]))
#Taking hue
plt.show()
```

```python
#Dropping the outliers by setting a cap on Age and income.
data = data[(data["Age"]<90)]
data = data[(data["Income"]<600000)]
print("The total number of data-points after removing the outliers are:", len(data))
data.info()
```

```python
#Get list of categorical variables
s = (data.dtypes == 'object')
object_cols = list(s[s].index)
print("Categorical variables in the dataset:", object_cols)
```

```python
#Label Encoding the object dtypes.
LE=LabelEncoder()
for i in object_cols:
    data[i]=data[[i]].apply(LE.fit_transform)
```

```python
print("All features are now numerical")
data.info()
```

#CREATING A COPY OF DATA FOR SCALING AND PCA

```python
#Creating a copy of data
ds = data.copy()
# creating a subset of dataframe by dropping the features on deals accepted and promotions
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
ds = ds.drop(cols_del, axis=1)
```

```python
#Scaling
scaler = StandardScaler()
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
print("All features are now scaled")
```

```python
#Scaled data to be used for reducing the dimensionality
print("Dataframe to be used for further modelling:")
scaled_ds.head()
```

```python
#Initiating PCA to reduce dimentions aka features to 3
pca = PCA(n_components=3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
#A 3D Projection Of Data In The Reduced Dimension
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

```python
# Quick examination of elbow method to find numbers of clusters to make.
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_ds)
Elbow_M.show()
```

```python
#Initiating the Agglomerative Clustering model
AC = AgglomerativeClustering(n_clusters=4)
```

```python
# fit model and predict clusters
yhat_AC = AC.fit_predict(PCA_ds)
PCA_ds["Clusters"] = yhat_AC
```

#ADDING THE CLUSTERS FEATURE TO THE ORIGNAL DATAFRAME

```python
#Adding the Clusters feature to the orignal dataframe.
data["Clusters"]= yhat_AC
```

```python
# Define the palette used elsewhere in your project
pal = ["#682F2F", "#B9C0C9", "#9F8A78", "#F3AB60"]
# Knowing PCA_ds contains your x, y, z coordinates and cluster labels
clusters = PCA_ds["Clusters"].unique()

# Start the 3D plot figure
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d') # Use add_subplot for cleaner 3D initialization
ax.set_title("The Plot Of The Clusters")
ax.set_xlabel("PC1") # It's good practice to label the axes
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")

# Iterate through each cluster to plot points and add a legend label
for i, cluster_id in enumerate(clusters):
    # Select the data points belonging only to the current cluster_id
    cluster_data = PCA_ds[PCA_ds["Clusters"] == cluster_id]

    # Scatter plot this subset, assigning a specific color from the palette (pal)
    # The 'label' argument is what creates the legend entry
    ax.scatter(
        cluster_data["col1"], # knowing 'col1' is x
        cluster_data["col2"], # knowing 'col2' is y
        cluster_data["col3"], # knowing 'col3' is z
        s=40,
        color=pal[i],
        marker='o',
        label=f"Cluster {cluster_id}"
    )

# Add the legend to the plot
ax.legend(title="Customer Segments")
plt.show()
```

```python
# Add 'Spent' to the list of features for relative importance
new_features_with_spent = ["Customer_For", "Education", "Family_Size", "Spent"]

# Calculate cluster averages
cluster_avg = data.groupby('Clusters')[new_features_with_spent].mean()

# Calculate population averages
population_avg = data[new_features_with_spent].mean()

# Calculate relative importance for each cluster
# 1. Use .div() to divide the cluster average by the population average
# 2. Subtract 1 to express the difference as a percentage change (e.g., 0.5 means 50% higher)
relative_imp_spent = cluster_avg[new_features_with_spent].div(population_avg[new_features_with_spent]) - 1

# Visualize the results
plt.figure(figsize=(10, 4))
plt.title("Relative Importance: Demographic, Loyalty, and SPENDING")
sns.heatmap(data=relative_imp_spent, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.show()
```
